````
import numpy as np
import seaborn as sns #связано с pandas
import matplotlib.pyplot as plt
import pandas as pd
````

````
tnt = sns.load_dataset("titanic")
tnt
````

# **Зависимость платы за проезд от возраста**









````
plt.plot(tnt.age, tnt.fare, 'gs', alpha=0.1)
plt.grid(True)
plt.ylabel("Плата за проезд")
plt.xlabel("Возраст")
plt.title("Зависимость платы за проезд от возраста")
#plt.plot(tnt.sex,tnt.age, 'bo')
````

**Зависимость выживаемости от возраста**

````
plt.plot(tnt.age,tnt.alive, 'ro', alpha=0.1)
plt.grid(True)
plt.title("Зависимость выжили или нет от возраста")
````
